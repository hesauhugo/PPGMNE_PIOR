```python
import numpy as np 
import copy
n = 12
```

```python
grafo = [
    [1,2,6],
    [1,3,8],
    [1,4,5],
    [2,7,8],
    [2,3,9],
    [3,4,7],
    [3,7,20],
    [3,6,10],
    [4,5,22],
    [5,8,9],
    [6,7,7],
    [6,8,4],
    [7,8,7]
    
]
```

```python
def corrigeGrafo(grafo):
    g = grafo.copy()
    gg = grafo.copy()
    
    for u,v,w in g:
        gg.append([v,u,w])
    return gg
```

```python
def retornaDistancia(solucao,grafo):
    dist = 0
    for i in range(len(solucao)-1):
        d = -1
        for a in range(len(grafo)):
            if grafo[a][0] == solucao[i] and grafo[a][1] == solucao[i+1]:
                d=grafo[a][2]
                break
        if d == -1:
            dist += float('inf')
            break
        else:
            dist += d
            
    
    return dist
```

```python
grafo = corrigeGrafo(grafo)
```

```python
print(grafo)
```
