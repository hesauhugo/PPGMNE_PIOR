1) b) Encontre o caminho mínimo, pelo algoritmo de Floyd, entre os pontos 3 e 10, utilizando o Sub-Grafo formado pelos pontos 3,4,5,6,7,8,9 e 10.

```python

inf = float('inf')
D_ant = [[0,6,8,5,inf,inf,inf,inf],
         [6,0,9,inf,inf,inf,8,inf],
         [8,9,0,7,inf,10,20,inf],
         [5,inf,7,0,22,inf,inf,inf],
         [inf,inf,inf,22,0,inf,inf,9],
         [inf,inf,10,inf,inf,0,7,4],
         [inf,8,20,inf,inf,7,0,7],
         [inf,inf,inf,inf,9,4,7,0]]

P_ant = [[-1] * n for i in range(n)]

n = len(D_ant)

for i in range(n):
    for j in range(n):
        if (0 < D_ant[i][j] < inf ):
            P_ant[i][j]=i
            
D = [[0] * n for i in range(n)]
P = [[0] * n for i in range(n)]

for k in range(n):
    for i in range(n):
        for j in range(n):
            if D_ant[i][j] <= D_ant[i][k] + D_ant[k][j] :
                D[i][j] = D_ant[i][j]
                P[i][j] = P_ant[i][j]
            else:
                D[i][j] = D_ant[i][k] + D_ant[k][j]
                P[i][j] = P_ant[k][j]
    
    for i in range(n):
        for j in range(n):
            D_ant[i][j] = D[i][j]
            P_ant[i][j] = P[i][j]
            
for i in range(n):
    for j in range(n):
        print('Caminho ótimo de', i+1,'para',j+1)
        print('Custo',D[i][j])
        print('Trajeto:',end = '')
        C = [j+1]
        k = P[i][j]
        while k != -1:
            C.append(k+1)
            k=P[i][k]
        print(list(reversed(C)))    
```
