Encontre , através do algoritmo de Dijkstra, o menor caminnho entre os pontos 1 e 12. Redesenhe o grafo e escreva a sequencia de pontos avaliandaso de acordo com o algoritmo.

```python
grafo = [
    [1,2,10],
    [1,3,7],
    [2,4,4],
    [3,4,6],
    [3,5,8],
    [3,6,5],
    [4,9,8],
    [4,5,9],
    [5,9,20],
    [5,8,10],
    [5,6,7],
    [6,7,22],
    [7,10,9],
    [7,11,6],
    [8,9,7],
    [8,10,4],
    [9,10,7],
    [10,12,13],
    [11,12,11]
]
```

```python
def corrigeGrafo(grafo):
    g = grafo.copy()
    gg = grafo.copy()
    
    for u,v,w in g:
        gg.append([v,u,w])
    return gg

def menorDistancia(dist,T,V):
    menor = 999999
    min_index=0
    for v in range(V):
        if dist[v]<menor and T[v]==False:
            menor = dist[v]
            min_index = v
            
    return min_index 

def dijkstra(grafo,src,V):
    dist = [999999]*V
    dist[src] = 0
    T = [False] * V
    caminho = []
    for count in range(V):
        u=menorDistancia(dist,T,V)
        for v in range(V):
            for item in grafo:
                if (item[0]-1==u) and (item[1]-1)==v:
                    if item[2]>0 and T[v] ==False and dist[v] >dist[u]+item[2]:
                        dist[v]    = dist[u]+item[2]
                        print('u=',item[0],"v=",item[1],dist[v])
                        caminho.append([item[0],item[1]])
                     
                        
        T[u]=True
    return dist,caminho
    
    
```

```python
grafo=corrigeGrafo(grafo)
dist,caminho = dijkstra(grafo,0,12)
solucao=[]
fim = 12
inicio = 1
finaliza = True
while finaliza:
    for i in range(len(caminho)):
        if caminho[i][1]==fim:
            solucao.append(caminho[i][1])
            fim = caminho[i][0]
            if caminho[i][0] == inicio:
                finaliza = False
                solucao.append(caminho[i][0])
            break    
solucao = list(reversed(solucao))
print("solucao = ",solucao)
    
        
```
